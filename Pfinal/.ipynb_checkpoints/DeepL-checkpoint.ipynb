```python
import pandas as pd
import numpy as np
import joblib
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

```

```python
# Cargar el dataset (asegúrate de ajustar la ruta a tu archivo)
df = pd.read_csv('chicago_crimes.csv')

# Convertir las columnas de fechas al formato datetime especificando el formato
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y', errors='coerce')
df['Updated On'] = pd.to_datetime(df['Updated On'], format='%m/%d/%Y', errors='coerce')

# Llenar valores nulos para columnas no datetime con 0 y columnas datetime con un valor por defecto (por ejemplo, la fecha mínima)
df_non_datetime = df.select_dtypes(exclude=[np.datetime64])
df_datetime = df.select_dtypes(include=[np.datetime64])

df_non_datetime.fillna(0, inplace=True)
df_datetime.fillna(pd.Timestamp.min, inplace=True)

# Combinar nuevamente los DataFrames
df = pd.concat([df_non_datetime, df_datetime], axis=1)

# Mostrar tipos de datos para verificar
print(df.dtypes)


```

```python
# Inicializar y ajustar LabelEncoders
categorical_columns = ['Case Number', 'Block', 'IUCR', 'Primary Type', 'Description', 'Location Description', 'FBI Code', 'Location']
label_encoders = {col: LabelEncoder() for col in categorical_columns}
# Convertir todas las columnas categóricas a cadenas
for col in categorical_columns:
    df[col] = df[col].astype(str)
    df[col] = label_encoders[col].fit_transform(df[col])

# Guardar los LabelEncoders
for col in categorical_columns:
    joblib.dump(label_encoders[col], f'label_encoder_{col.lower().replace(" ", "_")}.pkl')

# Mostrar las primeras filas del DataFrame para verificar
print(df.head())


```

```python
# Extraer características de las fechas (año, mes, día) para 'Date' y 'Updated On'
df['Year_Date'] = df['Date'].dt.year
df['Month_Date'] = df['Date'].dt.month
df['Day_Date'] = df['Date'].dt.day

df['Year_Updated'] = df['Updated On'].dt.year
df['Month_Updated'] = df['Updated On'].dt.month
df['Day_Updated'] = df['Updated On'].dt.day

# Eliminar las columnas originales de fechas
df = df.drop(columns=['Date', 'Updated On'])

# Verificar la conversión
print(df.head())
```

```python
# Seleccionar características y objetivo
features = df.drop(columns=['Domestic'])
target = df['Domestic']

# Dividir en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
# Escalar los datos
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Entrenar el modelo
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train_scaled, y_train)

# Hacer predicciones
y_pred = model.predict(X_test_scaled)

# Calcular métricas de evaluación
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print(f'Accuracy: {accuracy}')
print('Classification Report:')
print(report)
print('Confusion Matrix:')
print(conf_matrix)

#Guardar el modelo entrenado
joblib.dump(model, 'random_forest_model.pkl')
joblib.dump(scaler, 'scaler.pkl')
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
import joblib

# Cargar el dataset
df = pd.read_csv('chicago_crimes.csv')

# Convertir las columnas de fechas al formato datetime especificando el formato
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y', errors='coerce')
df['Updated On'] = pd.to_datetime(df['Updated On'], format='%m/%d/%Y', errors='coerce')

# Llenar valores nulos para columnas no datetime con 0 y columnas datetime con un valor por defecto
df_non_datetime = df.select_dtypes(exclude=[np.datetime64])
df_datetime = df.select_dtypes(include=[np.datetime64])

df_non_datetime.fillna(0, inplace=True)
df_datetime.fillna(pd.Timestamp.min, inplace=True)

# Combinar nuevamente los DataFrames
df = pd.concat([df_non_datetime, df_datetime], axis=1)

# Inicializar y ajustar LabelEncoders
categorical_columns = ['Case Number', 'Block', 'IUCR', 'Primary Type', 'Description', 'Location Description', 'FBI Code', 'Location']
label_encoders = {col: LabelEncoder() for col in categorical_columns}

for col in categorical_columns:
    df[col] = df[col].astype(str)
    df[col] = label_encoders[col].fit_transform(df[col])

# Guardar los LabelEncoders
for col in categorical_columns:
    joblib.dump(label_encoders[col], f'label_encoder_{col.lower().replace(" ", "_")}.pkl')

# Extraer características de las fechas
df['Year_Date'] = df['Date'].dt.year
df['Month_Date'] = df['Date'].dt.month
df['Day_Date'] = df['Date'].dt.day

df['Year_Updated'] = df['Updated On'].dt.year
df['Month_Updated'] = df['Updated On'].dt.month
df['Day_Updated'] = df['Updated On'].dt.day

# Eliminar las columnas originales de fechas
df = df.drop(columns=['Date', 'Updated On'])

# Definir características y objetivo
features = df.drop(columns=['Domestic'])
target = df['Domestic']

# Dividir en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Escalar los datos
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Entrenar el modelo
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train_scaled, y_train)

# Hacer predicciones
y_pred = model.predict(X_test_scaled)

# Calcular métricas de evaluación
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print(f'Accuracy: {accuracy}')
print('Classification Report:')
print(report)
print('Confusion Matrix:')
print(conf_matrix)

# Guardar el modelo y el scaler
joblib.dump(model, 'random_forest_model.pkl')
joblib.dump(scaler, 'scaler.pkl')
```
